### Varredadura dos parâmetros

Para ajudar no chute dos parâmetros vou fazer uma varredura pra achar a ordem de grandeza das regras

````python
def logGridSearch(support_ths, confidence_ths, dataset):    
    for th_it in support_ths:
#         calculate frequent itemsets
        frequent_itemsets = apriori(dataset, min_support=th_it, use_colnames=True, max_len=2)
        num_itemsets = len(frequent_itemsets["itemsets"])
#         generate rules
        for th_rl in confidence_ths:
            rules = []
            if num_itemsets > 0:
                rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=th_rl)
            num_rules = len(rules)
            print("( support_min: %f, confidence_min: %f ) #rules =  %i" % (th_it, th_rl, num_rules))
````

### calculando o chisquare para essas regras

Podemos utilizar uma função do pacote scipy, mas para isso, para cada linha do dataset de rules, teremos que calcular a tabela de contigênca. A documentação da função tem o seguinte exemlo:

```python
obs = np.array([[10, 10, 20], [20, 20, 20]])
chi2_contingency(obs)
```

````python
def selectComplementByFilm(dataset, film):
    wihFilm = dataset[dataset[film] == True]
    notWihFilm = dataset[dataset[film] == False]
    return wihFilm, notWihFilm
````

````python
def createContigencyTable(dataset, filmA, filmB):
    dfA, dfnA = selectComplementByFilm(dataset,filmA)
    dfAB, dfAnB = selectComplementByFilm(dfA, filmB)
    dfnAB, dfnAnB = selectComplementByFilm(dfnA, filmB)
    return np.array([[len(dfAB), len(dfAnB)], [len(dfnAB), len(dfnAnB)]])
````

````python
def calculateChiSquareForFule(df_pivoted, r):
    ant = str(r["antecedents"])
    ant = re.search("\\{'(.*)'\\}", ant)[1]
    des = str(r["consequents"])
    des = re.search("\\{'(.*)'\\}", des)[1]
    table = createContigencyTable(df_pivoted, ant, des)
    pvalue = chi2_contingency(table)[1]
    return pvalue
````

````python
rules_strong['chi2-pvalue'] = rules_strong.apply(lambda r: calculateChiSquareForFule(df_pivoted["id"], r), axis=1)
# df_pivoted["id"].apply(lambda x: x, axis=1)
````

````python
print("Números de regras que falham ao teste de independência com confiança .001", (rules_strong['chi2-pvalue'] > .001).sum())
````

### Export data to .csv

````python
def createOutFileRuleFromat(r):
    ant = str(r["antecedents"])
    ant = re.search("\\{'(.*)'\\}", ant)[1]
    des = str(r["consequents"])
    des = re.search("\\{'(.*)'\\}", des)[1]
    return ant + ' => ' + des
````

````python
def exportRulesToTxt(rules):
#     Regra | suporte | confiança | lift (B,C) e (B,!C) | chi-square...
    out = pd.DataFrame()
    out["regra"] = rules.apply(createOutFileRuleFromat, axis=1)
    out['suporte'] = rules['support']
    out['confiança'] = rules['confidence']
    out['lift (A=>B)'] = rules['lift']
    out['chi-square (pvalue)'] = rules['chi2-pvalue']
    out = out.sort_values(["suporte", "confiança", 'lift (A=>B)'])
    return out
````

````python
out = exportRulesToTxt(rules_strong)
````
